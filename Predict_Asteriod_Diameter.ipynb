## Eccentricity
The eccentricity of ellipse is less than 1, and greater than 0. The eccentricity of a ellipse helps us to understand how circular it is with reference to a circle. Eccentricity also measures the ovalness of the ellipse and eccentricity close to one refers to high degree of ovalness.

Eccentricity is basically the ratio of the distances of a point on the ellipse from the focus, and from the directrix.

## argument of perihelion
The orientation of an elliptical orbit can be specified by three orbital elements: the inclination, the ascending node and the argument of perihelion. Suppose we have an elliptical orbit with eccentricity, e and semi-major axis, a. The perihelion is the point of closest approach between the orbiting body (e.g. a planet) and the focus (e.g. the Sun lies at one focus of a planetary orbit, the other focus is empty).

If we rotate the axis of the orbit around the focus, then the rotation angle is the argument of perihelion (ω).
An argument of perihelion of 0° means that the body will be at perihelion right when it crosses the ecliptic from South to North. An argument of perihelion of 90° means that the body will be at perihelion at its Northmost distance from the ecliptic.
